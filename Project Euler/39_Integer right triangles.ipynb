```python
cnt = [0] * 1001
for a in range(1,1001) :
    for b in range(a,1001) :
        if a + b > 1000 : break
        for c in range(b,1001) :
            if a + b + c > 1000 : break 
            
            if a**2 + b**2 == c**2 :
                cnt[a+b+c]+= 1
                
cnt
```
