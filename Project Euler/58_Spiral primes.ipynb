```python
def is_prime(n) :
    ck = int(n**0.5)
    for d in range(2, ck+1) :
        if n % d == 0 : return False
    return True

p_cnt = 0
n = 1
while 1 :
    if is_prime(4*n**2-2*n+1) : p_cnt+= 1
    if is_prime(4*n**2+1) : p_cnt += 1
    if is_prime(4*n**2+2*n+1) : p_cnt += 1
    
    if p_cnt / (4*n) <= 0.1 :
        print(2*n+1)
        break

    n += 1
    
```
