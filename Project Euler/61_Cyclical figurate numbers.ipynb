```python
def cal3(n) :
    return n*(n+1)//2

def cal4(n) :
    return n*n

def cal5(n) :
    return n*(3*n-1)//2

def cal6(n) :
    return n*(2*n-1)

def cal7(n) :
    return n*(5*n-3)//2

def cal8(n) :
    return n*(3*n-2)

def f(cal) : 
    n = 1
    while(True) :
        x = cal(n)
        if len(str(x)) == 4 : 
            yield x
        n+=1
        if len(str(x)) > 4 :
            return  

arr = []
cal = [cal3, cal4, cal5, cal6, cal7, cal8]

for i in range(6) :
    arr.append(list(f(cal[i])))
```
