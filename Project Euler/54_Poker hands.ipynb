```python
f = open('poker.txt', 'r')
s = f.read().split('\n')
s
```

```python
rank = {'2' : 0, '3' : 1, '4' : 2, '5' : 3, '6' : 4, '7' : 5,
        '8' : 6, '9' : 7, 'T' : 8, 'J' : 9, 'Q' : 10, 'K' : 11, 'A' : 12}
def score(a) :
    cnt = {'S' : 0,'C' : 0,'H' : 0,'D' : 0}
    cnt_n = [0 for _ in range(13)]
    a.sort(key = lambda x : rank[x[0]])
    
    for card in a :
        cnt[card[1]] += 1
        cnt_n[rank[card[0]]] += 1

    sq = True
    for i in range(1,5) :
        if rank[a[i][0]] - rank[a[i-1][0]] != 1 :
            sq = False
            break
        
    if max(cnt.values()) == 5 :
        if sq :
            if a[0][0] == 'T' : return 9*13
            return 8*13 + rank[a[0][0]]
        return 5*13 + rank[a[-1][0]]
    if sq : return 4*13 + rank[a[0][0]]
    
    if max(cnt_n) == 4 : return 7*13 + cnt_n.index(4)
    if max(cnt_n) == 3 :
        c = cnt_n.index(3)
        cnt_n.sort(reverse=True)
        if cnt_n[1] == 2 :
            return 6*13 + c
        return 3*13 + c
    if max(cnt_n) == 2 :
        c = cnt_n.index(2)
        cnt_n.sort(reverse=True)
        if cnt_n[1] == 2 :
            return 2*13 + c
        return 1*13 + c
    return 0

def tie(a,b) :
    a.sort(key = lambda x : rank[x[0]], reverse = True)
    b.sort(key = lambda x : rank[x[0]], reverse = True)
    
    for i in range(5) :
        if rank[a[i][0]] > rank[b[i][0]] : return True
        elif rank[a[i][0]] < rank[b[i][0]] : return False
    return True
    

ans = 0

for game in s[:-1] :
    arr = game.split(' ')
    a = arr[:5]
    b = arr[5:]
    
    a_score = score(a)
    b_score = score(b)
    
    if a_score > b_score : ans += 1
    elif a_score == b_score :
        if tie(a,b) : ans += 1
        
print(ans)
    
    
```
