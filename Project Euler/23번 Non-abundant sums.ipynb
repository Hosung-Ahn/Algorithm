```python
def divisor_sum(n) :
    ret = 0 
    for d in range(1, n//2 + 1) :
        if n % d == 0 :
            ret += d
    return ret

abundant_lst = []

for n in range(28124) :
    if n < divisor_sum(n) :
        abundant_lst.append(n)
        
abundant_lst
```

```python
two_sum_abundant_lst = set()
for n1 in abundant_lst :
    if n1 > 28123 : break
        
    for n2 in abundant_lst :
        two_sum = n1 + n2
        if two_sum > 28123 : break
        two_sum_abundant_lst.add(two_sum)
        
two_sum_abundant_lst
```

```python
sum(two_sum_abundant_lst)
```
