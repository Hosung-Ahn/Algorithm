```python
from collections import defaultdict

MAX = 10_000
Prime = [True] * MAX
check = int(MAX**0.5)
Prime[0] = False 
Prime[1] = False

for i in range(2,check+1) :
    if Prime[i] : 
        for j in range(i*2, MAX, i) :
            Prime[j] = False
            
prime_lst = [ ]
for n in range(MAX) :
    if Prime[n] :
        prime_lst.append(n)
        
def is_prime(n) :
    ck = int(n**0.5)
    for d in range(2,ck+1) :
        if n % d == 0 : return False
    return True

def prime_factorization(n) : # n을 소인수 분해하여 dict 으로 return 한다.
    prime_factor = defaultdict(int)
    while (n > 1) :
        if is_prime(n) :
            prime_factor[n] += 1
            break
            
        for prime in prime_lst :
            if n % prime == 0 :
                prime_factor[prime] += 1
                n //= prime
                break
                    
    return prime_factor

def euler_func(n) :
    ret = n
    prime_factor = prime_factorization(ret)
    for p in prime_factor.keys() :
        ret //= p
        ret *= (p-1)
    return ret

def is_permutation(a, b) :
    if len(str(a)) != len(str(b)) : return False
    
    a_lst = [0 for _ in range(10)]
    b_lst = [0 for _ in range(10)]
    
    for c in str(a) :
        a_lst[int(c)] += 1
    for c in str(b) :
        b_lst[int(c)] += 1
        
    return a_lst == b_lst

min_x = 987654321
ans = 0

for n in range(2,10_000_000) :
    cn = euler_func(n)
    if is_permutation(n,cn) :
        tmp = n / cn
        if tmp < min_x :
            min_x = tmp
            ans = n
            
    if n % 100000 == 0 : print(n)
    
ans
```
