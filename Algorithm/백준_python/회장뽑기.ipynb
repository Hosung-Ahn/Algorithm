```python
import sys
from queue import deque
input = sys.stdin.readline
edges = [[] for _ in range(51)]

N = int(input())

a,b = map(int, input().split())
while(a != -1) :
    edges[a].append(b)
    edges[b].append(a)
    a,b = map(int, input().split())

def cal_score(a) :
    matched = [100 for _ in range(N+1)]
    q = deque()
    q.append(a)
    matched[a] = 0
    
    while(q) :
        cur= q.pop()
        for next in edges[cur] :
            if matched[next] <= matched[cur]+1 : continue
            matched[next] = matched[cur]+1
            q.append(next)
        
    ret = 0 
    for n in range(1,N+1) :
        if n == a : continue
        ret = max(ret, matched[n])
    return ret

score = 100
p_cnt = 0
p_lst = []

for p in range(1,N+1) :
    tmp = cal_score(p)
    
    if tmp > score : continue
    elif tmp < score :
        score = tmp
        p_cnt = 1
        p_lst = [p]
    else :
        p_cnt += 1
        p_lst.append(p)
        
print(score, p_cnt)
print(*p_lst)
```
