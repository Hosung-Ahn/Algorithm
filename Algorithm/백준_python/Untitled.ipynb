```python
n,x = map(int, input().split())
result = 987654321
def dfs(cur, cnt) :
    print(cur)
    global n
    global result
    
    if len(str(cur)) == n :
        result = min(result, cnt)
        return
    if len(str(cur)) > n :
        return
    
    lst = [0]*10
    for i in str(cur) :
        lst[int(i)] = 1
    for i in range(9,1,-1) :
        if lst[i] :
            dfs(cur * i, cnt+1)
dfs(x,0)
if x == 1 or result == 987654321: 
    print(-1)
else :
    print(result)
```
