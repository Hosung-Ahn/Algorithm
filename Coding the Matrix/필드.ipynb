```python
# ax + b = c 
def solve(a,b,c) :
    if a == 0 :
        print("error : a = 0, 0 can't be divisor")
        return
    return (c-b)/a

solve(10,5,30)
solve(10+5j, 5, 20)
```

```python
import matplotlib.pyplot as plt
import numpy as np

S = [2+2j, 3+2j, 1.75+1j, 2+1j, 2.25+1j, 2.5+1j, 2.75+1j, 3+1j, 3.25+1j ]

def drawPlot(S, x = (-5,5), y = (-5,5)) : 
    plt.scatter(np.real(S), np.imag(S))
    plt.xlim(x)
    plt.ylim(y)
    plt.axhline(y = 0, color = "Black")
    plt.axvline(x = 0, color = "Black")    

    plt.show()
```

```python
drawPlot(S)
S_scaled = [z/2 for z in S]
drawPlot(S_scaled)
```

```python
drawPlot(S)
S_truned90 = [z*1j for z in S]
drawPlot(S_truned90, (-5,5))
```

```python
import math

e = math.e
pi = math.pi
w = e**(2*pi*1j/20)
lst = [w**i for i in range(20)]

drawPlot(lst)
```

```python
S_rotated = [e**(1j*pi/4) * x for x in S]

drawPlot(S_rotated)
```

```python
S_complicated = [e**(1j*pi/4)/2 * x for x in S]
drawPlot(S_complicated)
```
